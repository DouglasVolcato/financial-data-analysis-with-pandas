```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.shift(periods= -1)
```

```python
aapl['lag1'] = aapl.shift(periods=1)
```

```python
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
aapl['pct_change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

```python
aapl.AAPL.resample('M').last().pct_change(periods=1).mul(100)
```

```python
close.head()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
ret.describe()
```

```python
summary = ret.describe().T.loc[:, ['mean', 'std']]
```

```python
summary.head()
```

```python
summary['mean'] = summary['mean']*252
summary['std'] = summary['std']*np.sqrt(252)
```

```python
summary.head()
```

```python
# risk and returns
summary.plot.scatter(x='std', y='mean', s=50)
plt.xlabel('Anual risk(std)')
plt.ylabel('Anual return(mean)')
plt.title('Rist/Return')

for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i, 'std']+0.002,summary.loc[i, 'mean']+0.002), size=15)

plt.show()
```

```python
# exponential averages
plt.figure(figsize=(12, 8))
aapl.ewm(span=10, adjust=False).mean().plot()
aapl.ewm(span=100, adjust=False).mean().plot()
aapl.ewm(span=200, adjust=False).mean().plot()
aapl.ewm(span=500, adjust=False).mean().plot()
plt.legend()
plt.show()
```
